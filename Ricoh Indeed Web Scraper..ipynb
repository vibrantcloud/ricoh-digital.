# Find the relevant parts of the review.

pass these into lists or JSON. 

```python
# Lets Break these down into parts and pass into a function to iterate on.
heading, job_title, rating, review,pro,con,location, date = [], [], [],[],[],[],[],[]
def get_ricoh_reviews(reviews,steps):
    
    """
    pass the url into the page
    scrape the relevant page
    iterate and pass into lists to account for nulls.
    """
    for i in range(0,reviews,steps):
        page = requests.get(f'https://www.indeed.co.uk/cmp/Ricoh/reviews?fcountry=GB&start={i}&sort=date_asc')
        soup = BeautifulSoup(page.text,'html.parser')
        container = soup.find_all('div',class_='cmp-review-container')
        time.sleep(0.5)
        for reviews in container:
            heading.append(reviews.find_all('div', class_='cmp-review-title'))
            job_title.append(reviews.find_all('span',class_='cmp-reviewer-job-title'))
            rating.append(reviews.find_all('div',class_='cmp-ratingNumber'))
            review.append(reviews.find_all('span',class_='cmp-review-text'))
            pro.append(reviews.find_all('div',class_='cmp-review-pro-text'))
            con.append(reviews.find_all('div',class_='cmp-review-con-text'))
            location.append(reviews.find_all('span',class_='cmp-reviewer-job-location'))
            date.append(reviews.find_all('span','cmp-review-date-created'))
            
    print('data_scraped')
        
    
    
    
    
    

```

```python
get_ricoh_reviews(reviews=250,steps=20)
```

# Iterate into a list.

```python
# create some lists. 
headings, job_titles, ratings, reviews_, pros, cons, locations, dates = [], [], [], [], [], [],[], []
```

```python
items = [heading, job_title, rating, review,pro,con,location, date]
```

```python
lists_ =  [headings, job_titles, ratings, reviews_, pros, cons, locations,dates]
```

```python
for i in range(len(items)):
    for elem in items[i]:
        try:
            lists_[i].append(elem[0].text)
        except IndexError:
            lists_[i].append(None)
```

```python
ricoh_reviews = pd.DataFrame({'Heading' : headings,
             'job_titles' : job_titles,
             'ratings' : ratings,
             'reviews' : reviews_,
             'pros' : pros,
             'cons' : cons,
             'locations' : locations,
             'Date' : dates})
```

```python
ricoh_reviews.to_csv('ricoh_indeed_review.csv',index=False)
```
